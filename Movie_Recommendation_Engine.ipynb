```
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```
df = pd.read_csv("movie_dataset.csv")
```

```
features = ['keywords', 'cast', 'genres', 'director']

for f in features:
  df[f] = df[f].fillna('')

def combineFeatures(row):
  ans = row['keywords'] + " " + row['cast'] + " " + row['genres'] + " " + row['director']
  return ans

df['combined_features'] = df.apply(combineFeatures, axis = 1)
```

```
countMatrix = cv.fit_transform(df['combined_features'])
cosineSimilarity = cosine_similarity(countMatrix)
```

```
movieName = input('Enter a movie name: ')
```

```
def getIndex(title):
  ans = df[df.title == title]['index'].values[0]
  return ans
```

```
movieIndex = getIndex(movieName)
similarMovies = enumerate(cosineSimilarity[movieIndex])
similarMovies = list(similarMovies)
```

```
sort = sorted(similarMovies, key = lambda x:x[1], reverse = True)
sort = sort[1:]
```

```
def extractTitle(index):
  ans = df[df.index == index]['title'].values[0]
  return ans
```

```
count = 0
names = []

for x in sort:
  name = extractTitle(x[0])
  names.append(name)
  count += 1

  if count == 5:
    break
```

```
print('Top 5 similar movies to ' + movieName + ' are the following: \n')
count = 1

for x in names:
  print(str(count) + '. ' + x + '\n')
  count += 1
```
